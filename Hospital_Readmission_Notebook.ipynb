```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
X.replace('?', pd.NA, inplace=True)
```

```python
X = X.drop(columns=['weight', 'payer_code', 'medical_specialty', 'max_glu_serum'])
```

```python
categorical_cols = X.select_dtypes(include='object').columns
X_encoded = X.copy()
label_encoders = {}
```

```python
for col in categorical_cols:
    le = LabelEncoder()
    X_encoded[col] = le.fit_transform(X_encoded[col].astype(str))
    label_encoders[col] = le
```

```python
y_encoded = LabelEncoder().fit_transform(y['readmitted'])
```

```python
imputer = SimpleImputer(strategy='most_frequent')
X_imputed = pd.DataFrame(imputer.fit_transform(X_encoded), columns=X_encoded.columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_imputed, y_encoded, test_size=0.2, random_state=42)
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_scaled, y_train)
y_pred_knn = knn.predict(X_test_scaled)
print("KNN Classification Report:\n", classification_report(y_test, y_pred_knn))
```

```python
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train_scaled, y_train)
y_pred_logreg = logreg.predict(X_test_scaled)
print("Logistic Regression Report:\n", classification_report(y_test, y_pred_logreg))
```

```python
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)  # Random Forest doesn't need scaled data
y_pred_rf = rf.predict(X_test)
print("Random Forest Report:\n", classification_report(y_test, y_pred_rf))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

# KNN
knn_params = {
    'n_neighbors': [3, 5, 7, 9],
}
knn_grid = GridSearchCV(KNeighborsClassifier(), knn_params, cv=3, scoring='accuracy', n_jobs=-1)
knn_grid.fit(X_train_scaled, y_train)

# Logistic Regression
logreg_params = {
    'C': [0.01, 0.1, 1, 10],
}
logreg_grid = GridSearchCV(LogisticRegression(max_iter=500), logreg_params, cv=3, scoring='accuracy', n_jobs=-1)
logreg_grid.fit(X_train_scaled, y_train)

# Random Forest
rf_params = {
    'n_estimators': [100, 150],
    'max_depth': [10, 5, None],
    'min_samples_split': [2, 5],
}
rf_grid = GridSearchCV(RandomForestClassifier(random_state=42), rf_params, cv=3, scoring='accuracy', n_jobs=-1)
rf_grid.fit(X_train, y_train)

# Neural Network (MLP)
mlp_params = {
    'hidden_layer_sizes': [(64,), (100,), (128,)],  # One hidden layer
    'alpha': [0.0001, 0.001],
}
mlp_grid = GridSearchCV(MLPClassifier(max_iter=300, random_state=42), mlp_params, cv=3, scoring='accuracy', n_jobs=-1)
mlp_grid.fit(X_train_scaled, y_train)
```

```python
print("🔍 Best KNN Parameters:", knn_grid.best_params_)
print("🔍 Best Logistic Regression Parameters:", logreg_grid.best_params_)
print("🔍 Best Random Forest Parameters:", rf_grid.best_params_)
print("🔍 Best Neural Network (MLP) Parameters:", mlp_grid.best_params_)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve

# Define models
models = {
    'KNN': KNeighborsClassifier(n_neighbors=9),
    'Logistic Regression': LogisticRegression(C=0.1, max_iter=500),
    'Random Forest': RandomForestClassifier(n_estimators=500, max_depth=15, random_state=42),
    'Neural Network': MLPClassifier(hidden_layer_sizes=(64,), alpha=0.0001, max_iter=300, random_state=42)
}

# Use your real data: X_train_scaled or X_train (scaled if needed)
X_used = X_train_scaled  # or X_train if no scaling needed
y_used = y_train

# Plot
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.ravel()

train_sizes = np.linspace(0.1, 1.0, 5)

for i, (name, model) in enumerate(models.items()):
    X_model = X_used if name != 'Random Forest' else X_train  # RF doesn't need scaling

    train_sizes_out, train_scores, test_scores = learning_curve(
        model, X_model, y_used,
        train_sizes=train_sizes, cv=3, scoring='accuracy', n_jobs=-1
    )

    axes[i].plot(train_sizes_out, train_scores.mean(axis=1), 'o-', label='Train')
    axes[i].plot(train_sizes_out, test_scores.mean(axis=1), 'o-', label='Validation')
    axes[i].set_title(f"Learning Curve: {name}")
    axes[i].set_xlabel("Training Set Size")
    axes[i].set_ylabel("Accuracy")
    axes[i].legend()
    axes[i].set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Identify misclassified indices
misclassified = {
    'KNN': set(np.where(y_test != y_pred_knn)[0]),
    'LogReg': set(np.where(y_test != y_pred_logreg)[0]),
    'Random Forest': set(np.where(y_test != y_pred_rf)[0]),
    'Neural Net': set(np.where(y_test != y_pred_mlp)[0])
}

# Step 2: Create empty DataFrame for directional overlap (% of A’s errors also made by B)
shared_error_pct = pd.DataFrame(index=misclassified.keys(), columns=misclassified.keys())

# Step 3: Fill in the matrix
for model_a in misclassified:
    for model_b in misclassified:
        errors_a = misclassified[model_a]
        errors_b = misclassified[model_b]
        if len(errors_a) == 0:
            shared_error_pct.loc[model_a, model_b] = 0.0
        else:
            shared = errors_a.intersection(errors_b)
            shared_error_pct.loc[model_a, model_b] = round(100 * len(shared) / len(errors_a), 1)

shared_error_pct = shared_error_pct.astype(float)

# Step 4: Print the table
print("🔁 Percentage of Model A's Errors Also Made by Model B:")
print(shared_error_pct)

# Step 5: Plot heatmap
plt.figure(figsize=(7, 6))
sns.heatmap(shared_error_pct, annot=True, cmap="Blues", fmt=".1f")
plt.title("Percentage of Model A's Errors Also Made by Model B")
plt.xlabel("Model B")
plt.ylabel("Model A")
plt.tight_layout()
plt.show()
```
